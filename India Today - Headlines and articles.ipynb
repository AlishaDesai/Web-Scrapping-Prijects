# India Today - Headlines and articles

```python
def indiatoday(uri):
    from bs4 import BeautifulSoup
    import requests
    #url = "http://"+uri
    url = uri
    data = requests.get(url)
    soup = BeautifulSoup(data.content)
    
    title = ""
    for i in soup.find_all("h1",itemprop={"headline"}):
        title = i.text
        
    summary = ""
    for i in soup.find_all("div",class_={"story-right"}):
        summary = summary + str(i.text)
        
    article = ""
    for i in soup.find_all("div",class_={"story-kicker"}):
        article = article + str(i.text)
        
    url_lengthth=len(url)
    date = url[(url_lengthth-10):url_lengthth]
    
    return title,summary,article,date
```

```python
def findurls_inapage(baseurl):
    from bs4 import BeautifulSoup
    import requests
    url = baseurl
    data = requests.get(url)
    soup = BeautifulSoup(data.content)
    urlsinapage = []
    for i in soup.find_all("div",class_={"view-content"}):
        for j in i.find_all("div",class_={"catagory-listing"}):
            for k in j.find_all("div",class_={"detail"}):
                for l in j.find_all("a"):
                    link = l['href']
                    story_or_vid = link[7:12]
                    if (story_or_vid=="story"):
                        full_url="https://www.indiatoday.in"+(link)
                        urlsinapage.append(full_url)
    return urlsinapage
```

```python
baseurls = []
allurls = []

for i in range(1,20,1):
    url = "https://www.indiatoday.in/india?page="
    url = url + str(i)
    baseurls.append(url)
    
for i in baseurls:
    allurls.extend(findurls_inapage(i))
```

```python
titles=[]
summary=[]
articles=[]
dates=[]
for i in allurls:
    t,s,a,d = indiatoday(i)
    titles.append(t)
    summary.append(s)
    articles.append(a)
    dates.append(d)

import pandas as pd
Q = pd.DataFrame([dates,summary,titles,articles]).T
Q.columns=["dates","summary","titles","articles"]
```
